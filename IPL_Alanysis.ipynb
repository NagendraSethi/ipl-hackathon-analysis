# **3. City which hosted most number of seasons according to the data is**

```python
# Group the data by 'city' and 'season', and count the number of occurrences
city_season_counts = df.groupby(['city', 'season']).size().reset_index(name='season_count')

# Find the city with the maximum number of seasons
city_with_most_seasons = city_season_counts.groupby('city')['season_count'].sum()

print("City which hosted the most number of seasons:", city_with_most_seasons.sort_values(ascending=False))
```

# **8. Which statement holds true**
1.   Teams winning tosses and electing field first have won more number of matches.
2.   Teams winning tosses and electing bat first have won more number of matches.
3.   Both A and B
4.   None of the above


```python
field_wins = matches[matches['toss_winner'] == matches['winner']]['toss_decision'].value_counts().get('field', 0)
bat_wins = matches[matches['toss_winner'] == matches['winner']]['toss_decision'].value_counts().get('bat', 0)
print(field_wins, bat_wins)
if field_wins > bat_wins:
  print(f"Teams winning tosses and electing to field first have won more number of matches.")
else:
  print(f"Teams winning tosses and electing to bat first have won more number of matches.")
```

# **10. To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?**

```python
batsmen_data = df.groupby('batsman')['batsman_runs'].sum().reset_index()
top_10_batsmen = batsmen_data.nlargest(10, 'batsman_runs')

# Plotting the most consistent batsmen among top 10 run-getters
plt.figure(figsize=(10, 6))
sns.lineplot(x='batsman', y='batsman_runs', data=top_10_batsmen, marker='o')
plt.title('Top 10 Run-Getters')
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# **11. In 2019 which batsman scored highest number of runs by hitting 6's and 4's?**

```python
df_2019=df[df['season']==2019]
boundaries = df_2019[df_2019['batsman_runs'].isin([4, 6])]
boundaries
```

```python
runs_by_batsman=boundaries.groupby('batsman')['batsman_runs'].sum().reset_index()
runs_by_batsman
```

```python
highest_scorer=runs_by_batsman.nlargest(1, 'batsman_runs')
print("Batsman who scored the highest number of runs from 6's and 4's in 2019:")
print(highest_scorer[['batsman', 'batsman_runs']])
```

# **20. The player with the highest hard-hitting ability by sixes**

```python
sixes=df[df['batsman_runs']==6].groupby('batsman')['batsman_runs'].count().reset_index()
sixes.sort_values(by='batsman_runs', ascending=False)
```

```python
sixes.columns = ['batsman', 'six_count']
highest_six_hitter = sixes.nlargest(1, 'six_count')

print("Player with the highest hard-hitting ability:")
print(highest_six_hitter[['batsman', 'six_count']])
```

# **The player with the highest hard-hitting ability by 4 & 6**

```python
df['hard_hitting_ability'] = (df['batsman_runs'].isin([4, 6]) / df['ball']) * 100

highest_hard_hitting_player = df.groupby('batsman')['hard_hitting_ability'].mean().idxmax()

print(f"The player with the highest hard-hitting ability: {highest_hard_hitting_player}")
```
